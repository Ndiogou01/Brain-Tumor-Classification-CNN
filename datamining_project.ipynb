```python
import tensorflow as tf
```

Preprocessing the Training set

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,                  
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
 
training_set = train_datagen.flow_from_directory(
    '/home/sns/pythonworkspace2/PROJET_DATAMINING/archive/Training',
    target_size=(64, 64),
    batch_size=32,
    class_mode='categorical'
)
print(training_set.class_indices)
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_directory(
    '/home/sns/pythonworkspace2/PROJET_DATAMINING/archive/Testing',
    target_size=(64, 64),
    batch_size=32,
    class_mode='categorical'
)
print(test_set.class_indices)
```

```python
cnn = tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu',input_shape=[64,64,3]))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=64, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dropout(0.5))
```

```python
cnn.add(tf.keras.layers.Dense(units=4, activation='softmax'))
```

```python
cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics= ['accuracy'])
```

```python
cnn.fit(x=training_set, validation_data=test_set, epochs=50)
```

```python
import numpy as np
from keras.preprocessing import image

test_image = image.load_img('/home/sns/pythonworkspace2/PROJET_DATAMINING/data/meningioma/Te-me_0011.jpg', target_size=(64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis=0)
test_image = test_image / 255.0

result = cnn.predict(test_image)
classes = ['glioma', 'meningioma', 'notumor', 'pituitary']
prediction = classes[np.argmax(result)]
print(prediction)
```
